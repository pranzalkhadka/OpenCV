```python
#Histogram is mainly used to visualize distribution of pixel density
import cv2 as cv
import numpy as np
```

```python
img = cv.imread('/home/pranjal-ubuntu/Downloads/OpenCV/Images/cat2.jpg')
```

```python
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
cv.imshow("gray",gray)
cv.waitKey(0)
```

```python
cv.destroyWindow("gray")
```

```python
hist = cv.calcHist([gray],[0],None,[256],[0,256])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.title("Grayscale histogram")
plt.xlabel("bins")
plt.ylabel("Nuber of pixels")
plt.plot(hist)
```

```python
#This means that there are almost 45000 pixels that have intensity of 250
```

```python
#FOR COLOR HISTOGRAM
colors=('b','g','r')
for i,col in enumerate(colors):
    hist = cv.calcHist([img],[i],None,[256],[0,256])
    plt.plot(hist,color=col)
      
plt.show()
```
